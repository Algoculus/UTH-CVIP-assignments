```python
import cv2
import matplotlib.pyplot as plt
```

```python
original_image = cv2.imread('original_image.png')
original_image = cv2.cvtColor(original_image, cv2.COLOR_BGR2RGB)
```

```python
# p dng b lc trung bnh (Average Filter)
average = cv2.blur(original_image, (10, 10)) # Tham s (10, 10) l kch thc ca kernel (ca s lc), s cng ln th cng m

# Hin th nh gc v nh sau khi lc trung bnh
titles = ['Original Image', 'Average Blurred Image']
images = [original_image, average]

plt.figure(figsize=(10, 5))
for i in range(len(images)):
    plt.subplot(1, len(images), i + 1)
    plt.imshow(images[i])
    plt.title(titles[i])
    plt.axis('off')
```
