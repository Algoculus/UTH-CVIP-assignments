```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Đọc ảnh cho bilateral filter
image_bilateral = cv2.imread("./images/animal.jpg")
image_bilateral_rgb = cv2.cvtColor(image_bilateral, cv2.COLOR_BGR2RGB)
image_bilateral_gray = cv2.cvtColor(image_bilateral, cv2.COLOR_BGR2GRAY)

# Đọc ảnh cho median filter
image_median = cv2.imread("./images/median.jpg")
image_median_rgb = cv2.cvtColor(image_median, cv2.COLOR_BGR2RGB)
image_median_gray = cv2.cvtColor(image_median, cv2.COLOR_BGR2GRAY)
```

```python
# Áp dụng Bilateral Filter
bilateral_filtered = cv2.bilateralFilter(image_bilateral_gray, d=35, sigmaColor=150, sigmaSpace=150)

plt.imshow(bilateral_filtered, cmap='gray')
plt.title("Bilateral Filter")
plt.axis('off')
```

```python
# Áp dụng Median Filter
median_filtered = cv2.medianBlur(image_median_gray, 27)

plt.imshow(median_filtered, cmap='gray')
plt.title("Median Filter")
plt.axis('off')
```

```python
# Hiển thị so sánh ảnh gốc và ảnh sau khi lọc
plt.figure(figsize=(14, 10))

# Bilateral Filter
plt.subplot(2, 2, 1)
plt.title("Original Image (Bilateral)")
plt.imshow(image_bilateral_gray, cmap='gray')
plt.axis('off')

plt.subplot(2, 2, 2)
plt.title("Bilateral Filter")
plt.imshow(bilateral_filtered, cmap='gray')
plt.axis('off')

# Median Filter
plt.subplot(2, 2, 3)
plt.title("Original Image (Median)")
plt.imshow(image_median_gray, cmap='gray')
plt.axis('off')

plt.subplot(2, 2, 4)
plt.title("Median Filter")
plt.imshow(median_filtered, cmap='gray')
plt.axis('off')

plt.tight_layout()
plt.show()
```

```python
# Lưu các ảnh kết quả vào thư mục result
cv2.imwrite('./result/bilateral_original.jpg', image_bilateral_gray)
cv2.imwrite('./result/bilateral_filtered.jpg', bilateral_filtered)
cv2.imwrite('./result/median_original.jpg', image_median_gray)
cv2.imwrite('./result/median_filtered.jpg', median_filtered)
```
