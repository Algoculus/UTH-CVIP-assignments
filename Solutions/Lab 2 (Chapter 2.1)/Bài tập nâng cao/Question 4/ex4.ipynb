```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Đọc ảnh cho bilateral filter
image_bilateral = cv2.imread("./images/bilateral.png")
image_bilateral_rgb = cv2.cvtColor(image_bilateral, cv2.COLOR_BGR2RGB)
#image_bilateral_gray = cv2.cvtColor(image_bilateral, cv2.COLOR_BGR2GRAY)

# Đọc ảnh cho median filter
image_median = cv2.imread("./images/image.png")
image_median_rgb = cv2.cvtColor(image_median, cv2.COLOR_BGR2RGB)
image_median_gray = cv2.cvtColor(image_median, cv2.COLOR_BGR2GRAY)
```

```python
# Áp dụng Bilateral Filter
bilateral_filtered = cv2.bilateralFilter(image_bilateral_rgb, d=5, sigmaColor=75, sigmaSpace=75)

plt.imshow(bilateral_filtered)
plt.title("Bilateral Filter")
plt.axis('off')
```

```python
# Áp dụng Median Filter
median_filtered = cv2.medianBlur(image_median_gray, 7)

plt.imshow(median_filtered, cmap='gray')
plt.title("Median Filter")
plt.axis('off')
```

```python
# Hiển thị so sánh ảnh gốc và ảnh sau khi lọc
plt.figure(figsize=(14, 10))

# Bilateral Filter
plt.subplot(2, 2, 1)
plt.title("Original Image (Bilateral)")
plt.imshow(image_bilateral_rgb)
plt.axis('off')

plt.subplot(2, 2, 2)
plt.title("Bilateral Filter")
plt.imshow(bilateral_filtered)
plt.axis('off')

# Median Filter
plt.subplot(2, 2, 3)
plt.title("Original Image (Median)")
plt.imshow(image_median_rgb, cmap='gray')
plt.axis('off')

plt.subplot(2, 2, 4)
plt.title("Median Filter")
plt.imshow(median_filtered, cmap='gray')
plt.axis('off')

plt.tight_layout()
plt.show()
```

```python
# Lưu các ảnh kết quả vào thư mục result
cv2.imwrite('./result/bilateral_original.jpg', cv2.cvtColor(image_bilateral_rgb, cv2.COLOR_RGB2BGR))
cv2.imwrite('./result/bilateral_filtered.jpg', cv2.cvtColor(bilateral_filtered, cv2.COLOR_RGB2BGR))
cv2.imwrite('./result/median_original.jpg', image_median_rgb)
cv2.imwrite('./result/median_filtered.jpg', median_filtered)
```
