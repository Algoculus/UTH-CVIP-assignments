```python
import cv2
import matplotlib.pyplot as plt
```

```python
gaussian_noise_image = cv2.imread('gaussian_noise_image.png')
gaussian_noise_image = cv2.cvtColor(gaussian_noise_image, cv2.COLOR_BGR2RGB)
```

```python
# p dng b lc Gaussian (Gaussian Filter)
# Tham s (11, 11) l kch thc kernel (phi l s l, v d: 3x3, 5x5, 7x7)
# Tham s 0 l  lch chun (sigmaX),  0 th OpenCV t tnh da trn kch thc kernel
gaussian = cv2.GaussianBlur(gaussian_noise_image, (11, 11), 0)

# Hin th nh gc v nh sau khi lc Gaussian
titles = ['Original Image', 'Gaussian Blurred Image']
images = [gaussian_noise_image, gaussian]

plt.figure(figsize=(10, 5))
for i in range(len(images)):
    plt.subplot(1, len(images), i + 1)
    plt.imshow(images[i])
    plt.title(titles[i])
    plt.axis('off')
```
