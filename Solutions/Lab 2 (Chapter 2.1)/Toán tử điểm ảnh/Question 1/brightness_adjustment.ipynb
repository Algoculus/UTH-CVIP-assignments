```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("image.jpg", cv2.IMREAD_GRAYSCALE)
if img is None:
    print("Error: Could not load image. Please check the file path.")
else:
    print(f"Image loaded successfully. Shape: {img.shape}")
```

```python
def change_brightness(image, value):
    result = image.astype(np.int16) + value
    result = np.clip(result, 0, 255)
    return result.astype(np.uint8)
```

```python
bright_img = change_brightness(img, 50)
dark_img   = change_brightness(img, -50)

plt.figure(figsize=(12,4))
plt.subplot(1,3,1); plt.title("Original"); plt.imshow(img, cmap='gray')
plt.subplot(1,3,2); plt.title("Brightness +50"); plt.imshow(bright_img, cmap='gray')
plt.subplot(1,3,3); plt.title("Brightness -50"); plt.imshow(dark_img, cmap='gray')
plt.show()
```
