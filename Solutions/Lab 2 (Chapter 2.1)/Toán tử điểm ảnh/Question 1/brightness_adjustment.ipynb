```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("image.jpg", cv2.IMREAD_GRAYSCALE)
if img is None:
    print("Error: Could not load image. Please check the file path.")
else:
    print(f"Image loaded successfully. Shape: {img.shape}")
```

```python
!# g(x,y)=f(x,y)+c => Tăng / giảm độ sáng ảnh
# Sáng / tối ảnh
# Ảnh xám thường là uint8 (0 -> 255)
def change_brightness(image, value):
    result = image.astype(np.int16) + value # tránh tràn số, Vì uint8 không có số âm và tràn modulo 256, int16 có thể <0 và >255
    result = np.clip(result, 0, 255) # <0 → 0, >255 → 255, Pixel hợp lệ phải nằm trong [0, 255]
    return result.astype(np.uint8) # unit 8 để chuẩn ảnh xám, tiết kiệm bộ nhớ, Tương thích OpenCV / matplotlib
```

```python
bright_img = change_brightness(img, 128)
dark_img   = change_brightness(img, -128)

plt.figure(figsize=(12,4))
plt.subplot(1,3,1); plt.title("Original"); plt.imshow(img, cmap='gray')
plt.subplot(1,3,2); plt.title("Brightness +128"); plt.imshow(bright_img, cmap='gray')
plt.subplot(1,3,3); plt.title("Brightness -128"); plt.imshow(dark_img, cmap='gray')
plt.show()
```
