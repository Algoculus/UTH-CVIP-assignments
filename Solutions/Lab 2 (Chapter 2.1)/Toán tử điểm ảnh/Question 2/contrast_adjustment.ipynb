```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("image.jpg", cv2.IMREAD_GRAYSCALE)
if img is None:
    print("Error: Could not load image. Please check the file path.")
else:
    print(f"Image loaded successfully. Shape: {img.shape}")
```

```python
# g(x,y)=α⋅f(x,y) => Tăng / giảm độ tương phản, α>1 tăng, 0<α<1 giảm, α=1 không đổi
# Rõ / mờ chi tiết
def change_contrast(image, alpha):
    result = image.astype(np.float32) * alpha # Dùng float32 để tránh tràn số khi nhân, vì uint8 chỉ từ 0 đến 255, float32 có thể vượt quá 255 và nhân có thể tạo ra số thập phân => tránh mất thông tin khi nhân
    result = np.clip(result, 0, 255) 
    return result.astype(np.uint8) 
```

```python
high_contrast = change_contrast(img, 1.5)
low_contrast  = change_contrast(img, 0.7)

plt.figure(figsize=(12,4))
plt.subplot(1,3,1); plt.title("Original"); plt.imshow(img, cmap='gray')
plt.subplot(1,3,2); plt.title("Contrast 1.5"); plt.imshow(high_contrast, cmap='gray')
plt.subplot(1,3,3); plt.title("Contrast 0.7"); plt.imshow(low_contrast, cmap='gray')
plt.show()
```
