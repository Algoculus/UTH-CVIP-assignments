```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("image.jpg", cv2.IMREAD_GRAYSCALE)
if img is None:
    print("Error: Could not load image. Please check the file path.")
else:
    print(f"Image loaded successfully. Shape: {img.shape}")
```

```python
def change_contrast(image, alpha):
    result = image.astype(np.float32) * alpha
    result = np.clip(result, 0, 255)
    return result.astype(np.uint8)
```

```python
high_contrast = change_contrast(img, 1.5)
low_contrast  = change_contrast(img, 0.7)

plt.figure(figsize=(12,4))
plt.subplot(1,3,1); plt.title("Original"); plt.imshow(img, cmap='gray')
plt.subplot(1,3,2); plt.title("Contrast 1.5"); plt.imshow(high_contrast, cmap='gray')
plt.subplot(1,3,3); plt.title("Contrast 0.1"); plt.imshow(low_contrast, cmap='gray')
plt.show()
```
