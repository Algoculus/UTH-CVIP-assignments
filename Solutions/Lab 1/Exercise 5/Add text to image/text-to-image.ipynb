```python
import cv2

img = cv2.imread("Requirements-lab1.png")

# Kiểm tra đọc ảnh thành công
print(img.shape)
```

```python
image = cv2.putText(
    img,
    "Exercise 5.2 - Lab 1",   # nội dung chữ
    (50, 50),                 # vị trí (x, y)
    cv2.FONT_HERSHEY_SIMPLEX, # font chữ
    0.6,                      # cỡ chữ
    (0, 0, 255),              # màu chữ (BGR) → đỏ
    2                         # độ dày
)

# Lưu ảnh đã có chữ
cv2.imwrite("Requirements-lab1-annotated.png", img)

cv2.imshow("Image with Text", image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
