```python
import cv2

img = cv2.imread("Requirements-lab1.png")

# Kiểm tra đọc ảnh thành công
print(img.shape)
```

```python
cv2.putText(
    img,
    "Exercise 5.2 - Lab 1",   # nội dung chữ
    (50, 50),                 # vị trí (x, y)
    cv2.FONT_HERSHEY_SIMPLEX, # font chữ
    0.6,                      # cỡ chữ
    (0, 0, 255),              # màu chữ (BGR) → đỏ
    2                         # độ dày
)
```

```python
from IPython.display import display
from PIL import Image

img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
display(Image.fromarray(img_rgb))
```

```python
cv2.imwrite("Requirements-lab1-annotated.png", img)
```
