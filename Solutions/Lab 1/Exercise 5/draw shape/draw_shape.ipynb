```python
# Vẽ các hình cơ bản lên ảnh
import cv2
import sys
import matplotlib.pyplot as plt
```

```python
# Đọc ảnh từ đường dẫn chỉ định (ảnh được đọc theo không gian màu BGR)
img = cv2.imread('images/original.jpg')

# Kiểm tra xem ảnh có được đọc thành công hay không
# Nếu đường dẫn sai hoặc OpenCV không đọc được ảnh thì img sẽ bằng None
if img is None:
    print("Lỗi: Không tìm thấy ảnh! Hãy kiểm tra lại đường dẫn.")
    sys.exit(1)  # Dừng chương trình với mã lỗi

# Lưu bản sao gốc của ảnh để so sánh sau này
img_original = img.copy()
```

```python
'''
Vẽ một đoạn thẳng lên ảnh.
 - img       : ảnh nền để vẽ
 - pt1       : điểm bắt đầu của đường thẳng (x, y)
 - pt2       : điểm kết thúc của đường thẳng (x, y)
 - color     : màu vẽ theo hệ BGR
 - thickness : độ dày đường vẽ (pixel)
 - lineType  : kiểu đường vẽ (mặc định cv2.LINE_8)
 - shift     : số bit dịch tọa độ (thường để 0)
'''
img = img_original.copy()

cv2.line(img, (0, 0), (200, 321), (255, 0, 0), 5)
cv2.line(img, (100, 0), (400, 321), (0, 255, 0), 10)

# Hiển thị kết quả tạm thời
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
cv2.imwrite('results/result_line.png', img)
plt.title("Draw line")
plt.axis("off")
plt.show()
```

```python

'''
Vẽ hình chữ nhật
- img       : ảnh nền
- pt1       : tọa độ góc trên bên trái
- pt2       : tọa độ góc dưới bên phải
- color     : màu vẽ (BGR)
- thickness : độ dày viền
              + thickness > 0 : vẽ viền
              + thickness = -1: tô kín hình chữ nhật
- lineType  : kiểu đường viền
- shift     : số bit dịch tọa độ
'''
img = img_original.copy()

# cv2.rectangle(img, (324, 420), (510, 128), (0, 255, 0), 3)
img_rectangle = cv2.rectangle(img, (224, 220), (310, 328), (0, 255, 0), 3)
img_rectangle = cv2.rectangle(img, (230, 200), (100, 100), (0, 0, 255), -1)

# Hiển thị kết quả tạm thời
plt.imshow(cv2.cvtColor(img_rectangle, cv2.COLOR_BGR2RGB))
cv2.imwrite('results/result_rectangle.png', img_rectangle)
plt.title("Draw rectangle")
plt.axis("off")
plt.show()
```

```python
'''Vẽ hình eclip
- img        : ảnh nền
- center     : tọa độ tâm elip
- axes       : độ dài hai bán trục (trục lớn, trục nhỏ)
- angle      : góc xoay của elip (độ)
- startAngle : góc bắt đầu của cung elip
- endAngle   : góc kết thúc của cung elip
- color      : màu vẽ (BGR hoặc giá trị xám)
- thickness  : độ dày đường vẽ
               + thickness = -1: tô kín elip
'''
img = img_original.copy()

cv2.ellipse(img, (120, 156), (100, 50), 0, 0, 240, 255, -1)
cv2.ellipse(img, (220, 156), (150, 75), 45, 0, 360, (0, 255, 0), 4)

# Hiển thị kết quả tạm thời
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
cv2.imwrite('results/result_ellipse.png', img)
plt.title("Draw ellipse")
plt.axis("off")
plt.show()
```

```python
'''Vẽ hình tròn
- img       : ảnh nền
- center    : tọa độ tâm hình tròn (x, y)
- radius    : bán kính hình tròn
- color     : màu vẽ (BGR)
- thickness : độ dày viền
              + thickness = -1: tô kín hình tròn
- lineType  : kiểu đường
- shift     : số bit dịch tọa độ
'''
img = img_original.copy()

cv2.circle(img, (400, 100), 50, (255, 0, 0), 3)
cv2.circle(img, (150, 100), 75, (0, 0, 255), -1)

# Hiển thị kết quả tạm thời
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
cv2.imwrite('results/result_circle.png', img)
plt.title("Draw circle")
plt.axis("off")
plt.show()
```
