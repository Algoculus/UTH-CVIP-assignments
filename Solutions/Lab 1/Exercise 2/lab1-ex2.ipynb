```python
import cv2
```

```python
image = cv2.imread("./mountain.jpg")
```

```python
# Check image if is loaded successfully
if image is None:
    print("Error: Counld not load image!")
else:
    print("Image loaded successfully!")
```

```python
# Display image
cv2.imshow("Image window", image)
cv2.waitKey(0)
cv2.destroyAllWindows()

for i in range (4):
    cv2.waitKey(1)
```

```python
# Write the image
image_new = image + 10
cv2.imwrite("./result.png", image_new)
```
