Import libraries

```python
import cv2
import matplotlib.pyplot as plt

# print(plt.colormaps())
```

Read original image

```python
bgr_image = cv2.imread("images/original_image.jpg")

if bgr_image is None:
    raise FileNotFoundError("Không tìm thấy ảnh images/original_image.jpg")

rgb_image = cv2.cvtColor(bgr_image, cv2.COLOR_BGR2RGB)
```

Identify common function

```python
def show_image(image, title="", cmap=None, size=(5, 5)):
    plt.figure(figsize=size)
    plt.imshow(image, cmap=cmap)
    plt.title(title)
    plt.axis("off")
    plt.show()

def convert_color(image, code):
    return cv2.cvtColor(image, code)
```

Show original image

```python
show_image(rgb_image, title="Original image (RGB)")
```

Convert BGR → Grayscale

```python
gray_image = convert_color(bgr_image, cv2.COLOR_BGR2GRAY)
show_image(gray_image, title="Grayscale", cmap="gray") 

# cmap là colormap, dùng để hiển thị ảnh xám, Pseudo-color, phân loại / đặc biệt
```

Convert BGR → HSV

```python
hsv_image = convert_color(bgr_image, cv2.COLOR_BGR2HSV)
show_image(hsv_image, title="HSV")
```

Convert BGR → LAB

```python
lab_image = convert_color(bgr_image, cv2.COLOR_BGR2LAB)
show_image(lab_image, title="LAB")
```

Function show images comparation

```python
def compare_images(images, titles, cmaps=None, size=(12, 4)):
    plt.figure(figsize=size)
    
    for i, img in enumerate(images):
        plt.subplot(1, len(images), i + 1)
        cmap = cmaps[i] if cmaps else None
        plt.imshow(img, cmap=cmap)
        plt.title(titles[i])
        plt.axis("off")
    
    plt.tight_layout() # tự căn chỉnh bố cục, tránh tràn lề + trồng tiêu đề
    plt.show()
```

```python
compare_images(
    images=[rgb_image, gray_image, hsv_image, lab_image],
    titles=["RGB", "Grayscale", "HSV", "LAB"],
    cmaps=[None, "gray", None, None]
)
```

Save result

```python
cv2.imwrite("results/gray.jpg", gray_image)
cv2.imwrite("results/hsv.jpg", hsv_image)
cv2.imwrite("results/lab.jpg", lab_image)
```
