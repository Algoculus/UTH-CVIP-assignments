```python
import cv2
import matplotlib.pyplot as plt
```

```python
image = cv2.imread("hcmc.jpg")
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
```

```python
# openCV automatically applies Gaussian + gradient + NMS + hysteresis
# output is binary image (uint8)
edges = cv2.Canny(gray, 100, 200) # 100 is the lower threshold, 200 is the upper threshold
```

```python
plt.figure(figsize=(10, 4))

plt.subplot(1, 2, 1)
plt.imshow(gray, cmap="gray")
plt.title("Grayscale Image")
plt.axis("off")

plt.subplot(1, 2, 2)
plt.imshow(edges, cmap="gray")
plt.title("Canny Edge Detection (OpenCV)")
plt.axis("off")

plt.tight_layout()
plt.show()
```
