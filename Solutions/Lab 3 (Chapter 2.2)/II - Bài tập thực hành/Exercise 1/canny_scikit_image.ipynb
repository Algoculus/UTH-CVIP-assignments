```python
import cv2
import matplotlib.pyplot as plt
from skimage import feature
```

```python
image = cv2.imread("hcmc.jpg")
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
```

```python
gray_norm = gray / 255.0 # normalize the image to 0-1
```

```python
edges = feature.canny(
    gray_norm,
    sigma=1.5,
    low_threshold=0.1,
    high_threshold=0.3
) # 0.1 is the lower threshold, 0.3 is the upper threshold, sigma is the standard deviation of the Gaussian filter
# Output is boolean image (True/False)
```

```python
plt.figure(figsize=(10, 4))

plt.subplot(1, 2, 1)
plt.imshow(gray, cmap="gray")
plt.title("Grayscale Image")
plt.axis("off")

plt.subplot(1, 2, 2)
plt.imshow(edges, cmap="gray")
plt.title("Canny Edge Detection (scikit-image)")
plt.axis("off")

plt.tight_layout()
plt.show()
```
